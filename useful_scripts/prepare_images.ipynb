```python
# Prepare the database images in a resolution of 1024x10224 and RGB

import os
from PIL import Image
import numpy as np

folder_path = '/content/drive/MyDrive/database'
file_names = os.listdir(folder_path)


for file_name in file_names:
  if file_name.endswith('.png'):
    # Load the image
    print(file_name)
    im_path = os.path.join(folder_path, file_name)
    im = Image.open(im_path)

    # Convert to RGB
    im_arr = np.array(im) * pow(2, 8) / pow(2, 16)
    im_8 = Image.fromarray(im_arr.astype(np.uint8))
    im_rgb = Image.merge('RGB', (im_8, im_8, im_8))

    # Resize to 1024
    im_rgb = im_rgb.resize((1024,1024))

    im_rgb.save('/content/drive/MyDrive/database_1024/' + file_name)
```
